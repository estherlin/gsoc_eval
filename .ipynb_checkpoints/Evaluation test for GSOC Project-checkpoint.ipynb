## Task: Acquire Data
The file name starts with an 18 digit number, which is the UNIX time in nanoseconds. Using the python datetime library and pytz, convert this number to a python datetime object in both UTC and CERN local time.

```python
# imports for this notebook
from datetime import datetime
from pytz import timezone
import re as re
import h5py
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# File name constants
FILENAME_REGEX = re.compile(
    r'(?P<time_stamp>\d{18}).*.h5',
    re.IGNORECASE
)
filename = '1541962108935000000_167_838.h5'

# Time zone constants
CERN_TIMEZONE = timezone('Europe/Zurich')
```

## Task: Time Conversions
The file name starts with an 18 digit number, which is the UNIX time in nanoseconds. Using the python datetime library and pytz, convert this number to a python datetime object in both UTC and CERN local time.

```python
# Convert 18 digit number to UTC and CERN time
match = FILENAME_REGEX.match(filename)
if match:
    unix_time = int(match.group('time_stamp'))
    utc_time = datetime.fromtimestamp(unix_time // 1000000000)
    cern_time = utc_time.astimezone(CERN_TIMEZONE)
```

## Task: Identify all datasets in the file
Use the h5py library to open the hdf file. The file is organized as a directory tree with groups and datasets. Write a program which explores all branches of the directory tree and identifies all of the datasets in the file. Your program should create a csv file which records the names of all of the groups and datasets, and includes the size, shape and type of  data in each dataset.  

The format of the csv will be as follows:

| name | size | shape| type |
| ---- | ---- | ---- | ---- |
| Group_n  |   |  | h5py.Group  |
| ...  |  ... |  ... | ...  |
| Dataset_n  | size_n  | shape_n | h5py.Dataset  |
| ...  |  ... |  ... | ...  |
with the elements being coloumns and each row corresponding to the information for one name. I will be adding None as the size and shape for Groups, since these attributes are not defined for Groups, only for Datasets. 

```python
# Define helper function to perform recursive traversal down directory tree
csv_data = []

def gather_info(name, obj):
    if isinstance(obj, h5py.Dataset):
        csv_data.append([name, obj.size, obj.shape, h5py.Dataset])
    elif isinstance(obj, h5py.Group):
        csv_data.append([name, None, None, h5py.Group])
```

```python
# Load in the provided file and gather the info on groups/datasets
f = h5py.File(filename,'r+')
f.visititems(gather_info)
#f.close()
```

```python
# Save gathered info on groups/datasets to csv
with open("%s_output.csv" % str(unix_time), "w") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerows(csv_data)
```

## Task: Process and save image
In the hdf file, there is 2D image data stored as a 1D array in the dataset called "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageData". The datasets "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageHeight" and "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageWidth" store information about the height and width of the image. Use numpy.reshape to convert the 1D array into a 2D image. Then, use scipy.signal.medfilt to filter the image. Finally, use matplotlib to display the image and save it as a png file.

```python
# Read in image data
img_name    = "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageData"
height_name = "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageHeight" 
width_name  = "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageWidth"

raw_img    = f[img_name][()]
img_height = f[height_name][()][0]
img_width  = f[width_name][()][0]
```

```python
# Resize img to convert from 1D to 2D
resized_img = np.resize(raw_img, (img_height, img_width))
plt.imshow(resized_img[:, :])
plt.title('Resized Streak Image')
plt.show()
```

```python
# Perform median filter on resized img
filter_img = signal.medfilt(resized_img)
plt.imshow(filter_img[:, :])
plt.title('Median Filtered Streak Image')
plt.savefig("%s_streak_img.png" % str(unix_time))
plt.show()
```
